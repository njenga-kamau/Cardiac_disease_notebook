```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

#Problem 1 

heart = pd.read_csv('heart.csv')

heart.head()

```

```python
heart.tail()
```

```python
#Problem 2 

#I just split the dataset manually 
X = heart.drop(columns = ['target'])
y = heart['target']


```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scale = scaler.fit_transform(X)
X_scale
```

```python
#Problem 3
#heart_data = pd.DataFrame(heart.data, columns=heart.feature_names)
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, random_state=0)
X_train,
```

```python
y_train
```

```python
#Problem 4
from sklearn.linear_model import LinearRegression


#X_train = X_train.dropna()
#y_train = y_train.dropna() 

logistic_regression = LogisticRegression(solver='lbfgs')
logistic_regression.fit(X_train, y_train)
```

```python
#Problem 5

predictions = logistic_regression.predict(X_test)

print("X_test predictions:", predictions)
```

```python
#Problem 6
accuracy = accuracy_score(y_test, predictions)
print(f'Logistic Regression Model Accuracy: {accuracy}')
```

```python
#Problem 7
data1 = np.array([0.8, -0.5, 0.2, 0.3, 1.2, -0.6, 0.9, -1.1, 1.5, -0.3, 0.7, 0.6, 1.1])  
data2 = np.array([-0.5, 0.3, -0.7, 0.9, -1.2, 0.5, 0.1, -0.8, 0.6, 0.2, -0.4, -0.9, -1.1])  

data1_scale = scaler.transform(data1.reshape(1, -1))
data2_scale = scaler.transform(data2.reshape(1, -1))

prediction_positive = logistic_regression.predict(data1_scale)
prediction_negative = logistic_regression.predict(data2_scale)

print("Prediction for one case (positive case):", prediction_negative[0])
print("Prediction for another case(negative case):", prediction_negative[0])
```
